```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np

class EmbeddingModel(nn.Module):
    def __init__(self, device, in_channel=5, out_channel=2):
        super(EmbeddingModel, self).__init__()
        self.conv = nn.Conv1d(in_channel, out_channel, kernel_size=3, padding=1)
        self.linear = nn.Linear(out_channel, out_channel)

        self.device = device
        self.to(self.device)
    
    def forward(self, x):
        if isinstance(x, np.ndarray): x = torch.Tensor(x).to(self.device)
        x = x.transpose(0, 1)  # Conv1d의 입력은 (N, C, L) 형태이므로 축을 변경
        x = self.conv(x)
        x = F.relu(x)
        x = x.transpose(0, 1)
        x = self.linear(x)
        return x
    
    def fit(self, X):
        pass  # 훈련 과정이 없으므로 무시
    
    def transform(self, X):
        with torch.no_grad():
            return self.forward(X).numpy()
    
    def to_numpy(self, tensor):
        if torch.is_tensor(tensor):
            return tensor.detach().cpu().numpy()
        else:
            return np.array(tensor)
```

```python
import spacy
from scipy.stats import entropy

# 문법적 다양성 점수를 계산하는 함수
def calculate_grammatical_scores(questions):
    nlp = spacy.load('en_core_web_sm')
    grammatical_scores = []
    for question in questions:
        if question.strip():
            doc = nlp(question)
            grammatical_score = 0
            for token in doc:
                if token.dep_ == 'ROOT':
                    grammatical_score += 1
                elif token.pos_ == 'VERB':
                    grammatical_score += 0.5
            grammatical_scores.append(grammatical_score)
    return grammatical_scores

# 정보 이득을 계산하는 함수
def calculate_information_gains(questions):
    information_gains = []
    for question in questions:
        if question.strip():
            # 문장을 단어로 분리
            words = question.split()
            # 각 단어의 빈도를 계산
            word_counts = {}
            for word in words:
                if word in word_counts:
                    word_counts[word] += 1
                else:
                    word_counts[word] = 1
            # 단어 빈도의 확률 분포 계산
            word_probabilities = [count / len(words) for count in word_counts.values()]
            # 정보 이득 계산
            information_gain = entropy(word_probabilities)
            information_gains.append(information_gain)
    return information_gains

# 문법적 다양성 점수 계산
grammatical_scores = calculate_grammatical_scores(questions)

# 정보 이득 계산
information_gains = calculate_information_gains(questions)

print(grammatical_scores)
print(information_gains)
```
